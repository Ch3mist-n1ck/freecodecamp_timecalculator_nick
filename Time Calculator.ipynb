```python
split_input=[]
def time_12hr(x):
    split_input=x.split()
    #print(split_input)
    
    global time_w
    if split_input[1]=="AM":
        time_w=-1
    elif split_input[1]=="PM":
        time_w=1
    else:
        print("AM/PM input has caused a problem, please try again using format 'x:xx XM'")
    #print(time_w)
    
    split_time_n=list(split_input[0])
    #print(split_time_n)
    min_time_n=[split_time_n[-2]+split_time_n[-1]]
    #print(min_time_n)
    if len(split_time_n)==4:
        hour_time_n=[split_time_n[0]]
    elif len(split_time_n)==5:
        hour_time_n=[split_time_n[0]+split_time_n[1]]
    else:
        print("too many digits used, please try again using format 'x:xx XM'")
    #print(hour_time_n)
    
    global hour_time
    global min_time
    try:
        hour_time=int(hour_time_n[0])
        #print(hour_time)
    except:
        print("only use numbers for the time, please try again using format 'x:xx XM'")
    hour_time=int(hour_time_n[0])
    try:
        min_time=int(min_time_n[0])
        #print(min_time)
    except:
        print("only use numbers for the time, please try again using format 'x:xx XM'")
    
def adder(y):
    split_input_a=y.split()
    #print(split_input_a)
    
    split_time_a=list(split_input_a[0])
    #print(split_time_a)
    min_time_a=[split_time_a[-2]+split_time_a[-1]]
    #print(min_time_a)
    for i in range(len(split_time_a)):
        try:
            int(split_time_a[i])
        except:
            digits=i
    hour_time_=""
    for i in range(digits):
        hour_time_+=split_time_a[i]
    hour_time_a=[hour_time_]
    #print(hour_time_a)
    
    try:
        hour_time_an=int(hour_time_a[0])
        #print(hour_time_an)
    except:
        print("only use numbers for the time, please try again using format 'x:xx XM'")
    try:
        min_time_an=int(min_time_a[0])
        #print(min_time_an)
    except:
        print("only use numbers for the time, please try again using format 'x:xx XM'")
    
    global new_hour
    global new_minute
    new_hour=hour_time+hour_time_an
    new_minute=min_time+min_time_an
    #print(new_hour)
    #print(new_minute)
    
def add_format():
    dec_min=[]
    tot_dec_min=""
    dec_hour=[]
    tot_dec_hour=""
    ad_hour=[]
    tot_ad_hour=""
    frac_min=new_minute/60
    #print(frac_min)
    split_frac_min=list(str(frac_min))
    #print(split_frac_min)
    #print(int(split_frac_min[1]))
    #
    #try:
    #    int(split_frac_min[2])
    #    double_digits=False
    #except:
    #    int(split_frac_min[2])
    #    double_digits=True
    #if                           will need this later, but the minutes at maximum will go to 118, which is a single digit
    global word_min
    for i in range(len(split_frac_min)-2):
        dec_min.append(split_frac_min[i+2])
        tot_dec_min+=dec_min[i]
    #print(dec_min)
    #print(tot_dec_min)
    added_min=int(60*float("0."+tot_dec_min))
    #print(added_min)
    if 1<added_min<10:
        word_min="0"+str(added_min)
    elif added_min==0:
        word_min="00"
    else:
        word_min=added_min
    if int(split_frac_min[0])==0:
        min_to_hour=0
    elif int(split_frac_min[0])==1:
        min_to_hour=1
    else:
        print("Error: Please use regular minutes format of 0-60")
    
    global word_hour
    frac_hour=(new_hour+min_to_hour)/12
    #print(frac_hour)
    split_frac_hour=list(str(frac_hour))
    #print(split_frac_hour)
    for i in range(len(split_frac_hour)):
        try:
            int(split_frac_hour[i])
        except:
            digits=i
    #print(digits)
    for i in range(len(split_frac_hour)-(digits+1)):
        dec_hour.append(split_frac_hour[i+(digits+1)])
        tot_dec_hour+=dec_hour[i]
    #print(dec_hour)
    #print(tot_dec_hour)
    #print("0."+tot_dec_hour)
    #print(12*float("0."+tot_dec_hour))
    added_hour=int(round(12*float("0."+tot_dec_hour)))
    #print(added_hour)
    if added_hour==0:
        word_hour="12"
    else:
        word_hour=added_hour
    for i in range(digits):
        ad_hour.append(split_frac_hour[i])
        tot_ad_hour+=ad_hour[i]
    #print(ad_hour)
    #print(tot_ad_hour)
    
    global new_time_w
    if int(tot_ad_hour)%2==1:
        new_time_w=-1*time_w
    elif int(tot_ad_hour)%2==0:
        new_time_w=1*time_w
    else:
        print("The hours have come out as fractional, please reenter without this")
    #print(new_time_w)
    
    if new_time_w==time_w:
        am_to_pm=0
    elif time_w==1:
        if new_time_w==1:
            am_to_pm=0
        elif new_time_w==-1:
            am_to_pm=1
    else:
        am_to_pm=0
    
    global day
    global days
    day=int(tot_ad_hour)/2+am_to_pm
    if int(day)==1:
        days="(next day)"
    elif int(day)==1.5:
        days="(next day)"
    elif int(day)>1:
        days="({} days later)".format(int(day))
    else:
        days=""
    
def day_number(z):
    global current_day
    if z==None:
        pass
    else:
        sep_day=list(z)
        #print(sep_day)
        if sep_day[0]=="m":
            current_day=0
        elif sep_day[0]=="M":
            current_day=0
        elif sep_day[0]=="t":
            if sep_day[1]=="u":
                current_day=1
            elif sep_day[1]=="U":
                current_day=1
            if sep_day[1]=="h":
                current_day=3
            elif sep_day[1]=="H":
                current_day=3
        elif sep_day[0]=="T":
            if sep_day[1]=="u":
                current_day=1
            elif sep_day[1]=="U":
                current_day=1
            if sep_day[1]=="h":
                current_day=3
            elif sep_day[1]=="H":
                current_day=3
        elif sep_day[0]=="w":
            current_day=2
        elif sep_day[0]=="W":
            current_day=2
        elif sep_day[0]=="f":
            current_day=4
        elif sep_day[0]=="F":
            current_day=4
        elif sep_day[0]=="s":
            if sep_day[1]=="a":
                current_day=5
            elif sep_day[1]=="A":
                current_day=5
            if sep_day[1]=="u":
                current_day=6
            elif sep_day[1]=="U":
                current_day=6
        elif sep_day[0]=="S":
            if sep_day[1]=="a":
                current_day=5
            elif sep_day[1]=="A":
                current_day=5
            if sep_day[1]=="u":
                current_day=6
            elif sep_day[1]=="U":
                current_day=6
        else:
            print("Please check spelling of the day")
        #print(current_day)
        global added_day
        dec_day=[]
        tot_dec_day=""
        add_day_number=current_day+round(day)
        #print(add_day_number)
        add_day=add_day_number/7
        split_day=list(str(add_day))
        for i in range(len(split_day)):
            try:
                int(split_day[i])
            except:
                digits=i
        #print(digits)
        for i in range(len(split_day)-(digits+1)):
            dec_day.append(split_day[i+(digits+1)])
            tot_dec_day+=dec_day[i]
        #print(dec_hour)
        #print(tot_dec_hour)
        added_day=int(7*float("0."+tot_dec_day))
        #print(added_day)
        global word_day
        if added_day==0:
            word_day="Monday"
        elif added_day==1:
            word_day="Tuesday"
        elif added_day==2:
            word_day="Wednesday"
        elif added_day==3:
            word_day="Thursday"
        elif added_day==4:
            word_day="Friday"
        elif added_day==5:
            word_day="Saturday"
        elif added_day==6:
            word_day="Sunday"

def printer(z):
    if z==None:
        if new_time_w==-1:
            new_time=str(word_hour)+":"+str(word_min)+" "+"AM"+" "+days
        elif new_time_w==1:
            new_time=str(word_hour)+":"+str(word_min)+" "+"PM"+" "+days
        print(new_time)
    else:
        if new_time_w==-1:
            new_time=str(word_hour)+":"+str(word_min)+" "+"AM"+" , "+str(word_day)+" "+days
        elif new_time_w==1:
            new_time=str(word_hour)+":"+str(word_min)+" "+"PM"+" , "+str(word_day)+" "+days
        print(new_time)
    
def add_time(x,y,z=None):
    time_12hr(x)
    adder(y)
    add_format()
    day_number(z)
    printer(z)
    
add_time("9:00 PM","168:00")
```
